```python
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
from tensorflow.python.keras.layers import Input
from tensorflow.python.keras.layers import Flatten
from tensorflow.python.keras.callbacks import ModelCheckpoint

from keras.applications.resnet import ResNet
from keras.preprocessing.image import image_utils
import matplotlib.pyplot as plt
print('Environment set')
```

```python
"""
step 1: load the image with image_utils
step 2: crop the image to select the interest of region or patch
step 3: resize the image to 224 * 224 as ResNet input size. Here we are not designing the model for various input image sizes.

Note: Do not re-size upfront with the image_utils.load_image as we need to select the patch region and then resize the image for ResNet model.
"""
image = image_utils.load_img('D:/IITJ_classes/class material/semester4/DeepLearning/assignments/utkarsh/COL828_Assignment_1/Train/images/A_1109_1.LEFT_CC.jpg')
plt.imshow(image)
plt.show()
w, h = image.size
print(w,h)
#interested region 92.33697891	68.47298145	640	532.6651001

image_cropped = image.crop((92.33697891, 68.47298145, 640, 532.6651001))
w_cropped, h_cropped = image_cropped.size
plt.imshow(image_cropped)
plt.show()
print(w_cropped, h_cropped)

image_resized = image_cropped.resize((224, 224))
w_resize, h_resize = image_resized.size
plt.imshow(image_resized)
plt.show()
print(w_resize, h_resize)
```

```python
base_model = ResNet(include_top = False, weights = "imagenet", input_shape = (224,224, 3))
base_model.save_weights()
model = Sequential()
model.add(base_model)
model.add(Flatten())
model.add(Dense(80, activation="softmax"))

for layer in base_model.layers:
    layer.trainable = False
model.load_weights('all_layers_freezed.h5')

for layer in base_model.layers[-26:]:
    layer.trainable = True
```
